```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import csv
from cycler import cycler
import matplotlib.cm as cm

%matplotlib inline
```

```python
CollDir='/afs/cern.ch/work/n/nbiancac/scratch0/IRIS/LHC_IW_model/Coll_settings/'
ResultDir='/afs/cern.ch/user/n/nbiancac/ln_work/scratch0/DELPHI_results/LHC/MD/Results/'
comment='prova'
filename=[
    'LHC_ft_6.5TeV_B1_2016_Fill-5130_20160729_Scan_3_open',
    'LHC_ft_6.5TeV_B1_2016_Fill-5130_20160729_Scan_3_closed',
    'LHC_ft_6.5TeV_B1_2016_Fill-5130_20160729_Scan_4_open',
    'LHC_ft_6.5TeV_B1_2016_Fill-5130_20160729_Scan_4_closed',
]

colors = cm.rainbow(np.linspace(0, 1, len(filename)))

fig = plt.figure(figsize=(8,10))
ax= fig.add_subplot(111)

for ind,fn in enumerate(filename):
    
    reader = csv.DictReader(open(CollDir+fn+'.txt'),delimiter='\t')
    names = []; gaps = []
    for row in reader:
        names.append(row['Name'])
        gaps.append(eval(row['Halfgap[m]'])*1e3)
    x_vec = np.array(gaps)
    
    ax.barh(np.arange(len(x_vec))+.2*ind, x_vec, .1, 
             align='center', edgecolor=colors[ind], facecolor = colors[ind], alpha = 1, linewidth = 2)
    
    
plt.yticks(np.arange(len(names)), names)
plt.ylim(0,len(names))
plt.xlim(0,15)
plt.xlabel('Half gap [mm]')
plt.grid()
lgd = plt.legend(filename,bbox_to_anchor=(1.05, 1.215))

plt.savefig(ResultDir+comment+'_gaps'+'.png', bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.savefig(ResultDir+comment+'_gaps'+'.pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')    
print ResultDir+comment+'_gaps'+'.png'
```
